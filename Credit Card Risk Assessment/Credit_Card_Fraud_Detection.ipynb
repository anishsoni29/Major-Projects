Importing the Dependencies

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading the dataset into a dataframe
credit_card = pd.read_csv('creditcard.csv')
```

```python
#printing the first 5 rows of the dataset
credit_card.head()
```

```python
credit_card.tail()
```

```python
#information about the dataset
credit_card.info()
```

```python
#checking the number of missing values in each column
credit_card.isnull().sum()
```

```python
#distribution of legit transaction & fraudulent transactions
credit_card['Class'].value_counts()
```

The dataset is highly unbalanaced

0 ---> Normal Transaction
1 ---> Fraudulent Transaction

```python
#separating the data for analysis
legit = credit_card[credit_card.Class ==0]
fraud = credit_card[credit_card.Class ==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card.groupby('Class').mean()
```

Under - Sampling

Build a sample dataset containing similar distribution of Normal transactions and Fraudulent Transactions

No. of Fraudulent Transactions --> 93

```python
legit_sample = legit.sample(n=93)
```

Concatinating two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud], axis =0)
```

```python
new_dataset.head()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.tail()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
X = new_dataset.drop(columns = 'Class', axis =1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into Training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, stratify = Y, random_state =2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression()
```

```python
#training the Logistic Regression Model with Training Data
model.fit(X_train, Y_train)
```

Model Evaluation

Accuracy Score

```python
#accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Accuracy on Training Data :' , training_data_accuracy)
```

```python
#accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy score on Test Data : ', test_data_accuracy)
```
